# Youtube video downloader

#### Import `pafy` & `youtube-dl`

```python
import pafy
import youtube_dl
```

#### Taking the link of the video as input

```python
link = input("Enter the video link : ")
```

#### Download the video at best resolution 

```python
video = pafy.new(link)

best = video.getbest()
print(best.resolution, best.extension)
best.download()
```
